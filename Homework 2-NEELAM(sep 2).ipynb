# Task 1

### The gravitational force between two masses at distance r is $F = G \\frac{m1m2}{r^2}$

### The gravitational potential of a mass at some distance is  ùëâ= G \\frac{m1}{r} ¬∂


### The gravitational intensity of a mass at some distance $g = G \\frac{m1}{r^2}$

```python
class Gravitation():
    def __init__(self,G,m1,m2,r):
        self.G=G
        self.m1=m1
        self.m2=m2
        
    def Force(self,r):
        F=self.G*self.m1*self.m2/r**2
        return F
    def potential(self,r):
        V=self.G*self.m1/r
        return V
    def intensity(self,r):
        E=self.G*self.m1/r**2
        return E
```

```python
G=6.67*10e-11
m1=5.57*10e-20 #in kg
m2=3.25*10e-17 #in kg
R=[] #empty list to collect distance r
F=[] #empty list to collect force F
V=[] #empty list to collect potential V
E=[] #empty list to collect intensity E
for r in range (1,3700): #for loop
    C1= Gravitation(G,m1,m2,r)
    f=C1.Force(r)
    v=C1.potential(r)
    e=C1.intensity(r)
    R.append(r)
    F.append(f)
    V.append(v)
    E.append(e)
    #print(r)
```

```python
import matplotlib.pyplot as plt #plt is arbitary
plt.figure(figsize = [20,20])
plt.subplot(5,6,7)
plt.plot(R,F,label='Force')
plt.plot(R,V,label='Potential')
plt.plot(R,E,label='Intensity')
plt.legend()
plt.xlabel(r)
plt.ylabel('F,V,E')
plt.subplot(5,5,5)
plt.semilogy(R,F,label='Force')
plt.semilogy(R,V,label='Potential')
plt.semilogy(R,E,label='Intensity')
plt.legend()
plt.xlabel(r)
plt.ylabel('lnF,lnV,lnE')
plt.subplot(5,5,4)
plt.loglog(R,F,label='force')
plt.loglog(R,V,label='potential')
plt.loglog(R,E,label='intensity')
plt.legend()
plt.xlabel('lnr')
plt.ylabel('lnF,lnV,lnE')
```

```python
class Comparision():
    def __init__(self,G,K,m1,m2,q1,q2,r):
        self.G=G
        self.K=K
        self.q1=q1
        self.q2=q2
        self.m1=m1
        self.m2=m2
        self.r=r
    def Gforce(self):
        F1=(self.G*self.m1*self.m2)/(self.r**2)
        return F1
    def Eforce(self):
        F2=(self.K*self.q1*self.q2)/(self.r**2)
        return F2
```

```python
C=Comparision(6.67e-11,9e9,9.1e-31,1.67e-27,1.6e-19,1.6e-19,5.29e-11)
g=C.Gforce()
e=C.Eforce()
if e>g:
   print("Electrostatic force is greater than Gravitataional force")
if e<g:
   print("Gravitational force is greater than Electrostatic force")
if e==g:
   print("Gravitational force is equal to Electrostatic force")
```

```python
class Variation:
    def __init__(self,G,K,m1,m2,q1,q2):
        self.G=G
        self.K=K
        self.m1=m1
        self.m2=m2
        self.q1=q1
        self.q2=q2
    def F1(self,r):
        F1=(self.G*self.m1*self.m2)/(r**2)
        return F1
    def F2(self,r):
        F2=(self.K*self.q1*self.q2)/(r**2)
        return F2
```

```python
V=Variation(6.67e-11,9e9,9.1e-31,1.67e-27,1.6e-19,1.6e-19)
X=[] #empty list to collect electrostatic force
R=[] #empty list to colect distance
S=[] #empty list to collect gravitational force
for r in range (1,900):
    g=V.F1(r)
    e=V.F2(r)
    X.append(r)
    R.append(g)
    S.append(e)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=[30,30])
plt.subplot(5,5,4)
plt.plot(R,X, label ="Electrostatic force")
plt.plot(X,S, label ="Gravitational force")
```
