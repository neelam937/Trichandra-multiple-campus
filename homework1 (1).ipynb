## Electrostatic Force between the charges

## The Coulumb's force between two charges $q_1$ and $q_2$ separated at distance $r$ is given as;

### $F= k\\{q_1q_2}/{r^2}$ where $k=9\\*10^9Nm^2/(C^2)$

## Write a code for function to calculate the Coulumb's force between two given charges at given distance.

## Use list to calculate the variation of Coulumb's force between two charges(as you like) place at a range of distance(choose range in your favour)

# Make a plot that shows the variation of Coulumb's force with distance using list.

# solution;

### Let the 2 charges be q1=5C and q2=6C

```python
#Function to calculate the Coulumb's force
k=9*10^9
def Force(q1,q2,r):
    return(k*q1*q2/(r^2))
```

```python
#Range of distance
F=[]
M=[]
q1=5
q2=6
for r in range(50,5000):
    F.append(Force(r,q1,q2))
    M.append(r)
```

```python
#Plot between distance and Coulumb's force
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(F,M)
```
